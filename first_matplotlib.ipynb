```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = [1,5,8,3,7]
b = [25,29,40,45,19]
```

## LINE CHART

```python
plt.plot(a,b)
```

```python
plt.plot(a,b,color='r')
```

```python
plt.plot(a,b,color='r',marker='d')
```

```python
plt.plot(a,b,color='r',marker='o')
```

```python
plt.plot(a,b,color='b',marker='o')
plt.grid(True)
```

```python
plt.plot(a,b,color='b',marker='o')
plt.grid(True,color='r')
```

```python
plt.plot(a,b,color='b',marker='o')
plt.grid(True,color='r')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Line Chart")
```

```python
plt.plot(a,b,color='b',marker='o')
plt.grid(True,color='r')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Line Chart")
plt.grid(color = 'green', linestyle = '--', linewidth = 1)
```

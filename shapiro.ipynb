```python
from scipy.stats importg shapiro
import seaborn as sns
```

```python
data = [10,11,15,19,20,13,18]
s_stat,p_value = shapiro(data)
print(s_stat)
print(p_value)
```

```python
alpha = 0.05
if(p_value<alpha):
    print("Data is not normally distributed, Rej null hypothesis")
else:
    print("Data is normally distributed, Unable to rej null hypothesis")
```

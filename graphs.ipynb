```python
f=pd.DataFrame({"Name":["Abc","def","ghi","jkl"],
                 "Age":[21,25,23,19],
                 "Weight":[70,65,55,35],
               "Height":[170,172,165,135]})
print(f)
```

```python
plt.scatter(f.Age, f.Weight, marker='o',color='r')
```

```python
plt.plot(f.Age, f.Weight, marker='o',color='r')
```

```python
plt.bar(f.Height, f.Weight,color='b')
```

```python
import seaborn as sns
data1=pd.DataFrame({'name':['a','b','c','d'],
                    'age':[21,22,28,25],
                    'height':[50,120,160,90]})
sns.barplot(x='name',y='age',data=data1)
```

```python
sns.barplot(x='age',y='height',data=data1)
```

```python
sns.barplot(x='age',y='height',data=data1,hue='height')
```

```python
sns.lineplot(x='age',y='height',data=data1,marker='o',color='black')
```

```python
sns.lineplot(x='age',y='height',data=data1,marker='o',hue='height',color='black')
```

```python
d=np.random.randn(1000)
sns.histplot(d,bins=500,kde=False,color='grey')
```

## A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram.

```python
d=np.random.randn(1000)
sns.histplot(d,bins=500,kde=True,color='grey')
```
